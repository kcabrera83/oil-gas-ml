```python
# Importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuracion de visualizacion
plt.style.use('seaborn-v0_8')
sns.set_palette('husl')
pd.set_option('display.max_columns', None)
```

```python
# Cargar datos (ejemplo con CSV)
# Reemplaza 'ruta/a/tu/dataset.csv' con la ruta real
try:
    df = pd.read_csv('../data/produccion_pozos.csv')
    print('Datos cargados exitosamente')
except FileNotFoundError:
    print('Archivo no encontrado. Creando dataset de ejemplo...')
    df = pd.DataFrame({
        'pozo_id': range(1, 101),
        'produccion_bpd': np.random.normal(500, 150, 100),
        'presion_psi': np.random.normal(3000, 500, 100),
        'temperatura_c': np.random.normal(80, 15, 100),
        'porosidad': np.random.uniform(0.1, 0.3, 100),
        'permeabilidad_md': np.random.exponential(100, 100),
        'tipo_fluido': np.random.choice(['petroleo', 'gas', 'agua'], 100)
    })
```

```python
# Vista inicial de los datos
print('Dimensiones del dataset:', df.shape)
display(df.head())
```

```python
# Informacion general
print('\nInformacion del DataFrame:')
df.info()

print('\nEstadisticas descriptivas:')
display(df.describe())
```

```python
# Analisis de valores nulos
null_counts = df.isnull().sum()
null_percent = (df.isnull().sum() / len(df) * 100).round(2)

null_df = pd.DataFrame({'Nulos': null_counts, '%': null_percent})
print('Valores nulos por columna:')
display(null_df[null_df['Nulos'] > 0] if null_df['Nulos'].any() else 'Sin valores nulos')
```

```python
# Visualizacion: Distribucion de produccion
plt.figure(figsize=(10, 6))
sns.histplot(df['produccion_bpd'], bins=30, kde=True)
plt.title('Distribucion de Produccion (BPD)')
plt.xlabel('Produccion (barriles por dia)')
plt.ylabel('Frecuencia')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Matriz de correlacion (solo numericass)
numeric_cols = df.select_dtypes(include=[np.number]).columns

plt.figure(figsize=(10, 8))
corr_matrix = df[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Matriz de Correlacion')
plt.tight_layout()
plt.show()
```

```python
# Produccion vs Presion (ejemplo de relacion)
if 'presion_psi' in df.columns and 'produccion_bpd' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='presion_psi', y='produccion_bpd', data=df, hue='tipo_fluido', alpha=0.7)
    plt.title('Produccion vs Presion del Pozo')
    plt.xlabel('Presion (PSI)')
    plt.ylabel('Produccion (BPD)')
    plt.legend(title='Tipo de Fluido')
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
```

```python
# Resumen final
print('RESUMEN DEL ANALISIS EXPLORATORIO')
print('=' * 50)
print(f'- Registros totales: {len(df)}')
print(f'- Columnas: {df.shape[1]}')
print(f'- Columnas numericas: {len(df.select_dtypes(include=[np.number]).columns)}')
print(f'- Columnas categoricas: {len(df.select_dtypes(include=["object"]).columns)}')
print(f'- Valores nulos totales: {df.isnull().sum().sum()}')
print(f'- Duplicados: {df.duplicated().sum()}')
print('=' * 50)
print('Analisis completado. Listo para siguiente fase.')
```
